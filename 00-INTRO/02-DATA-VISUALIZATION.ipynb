### Data Visulization with `ggplot2`
In this notebook we are going to learn about how to do data visualization using `R`.

#### What is ggplot2?
- `ggplot2` is part of the tidyverse and is based on The Grammar of Graphics.
- It builds plots layer by layer, starting with data and mapping, then adding geometries, scales, themes, labels, etc.

> https://ggplot2.tidyverse.org/

First we are going to load `ggplot2` for that we are going to use `pacman` as follows:

```r
pacman::p_load(pacman, ggplot2)
```

### 1. Scatter Plot - Sepal.Length vs Sepal.Width by Species
- Show relationship between two continuous variables, colored by species.

```r
ggplot(data = iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) +
  geom_point(size = 3) +
  labs(
    title = "Sepal Length vs Sepal Width",
    x = "Sepal Length (cm)",
    y = "Sepal Width (cm)",
    color = "Species"
  ) +
  theme_minimal()
```

### 2. Histogram - Distribution of Petal.Length
- See how a numeric variable is distributed.

```r
ggplot(iris, aes(x = Petal.Length, fill = Species)) +
  geom_histogram(binwidth = 0.5, alpha = 0.7, color = "black") +
  labs(
    title = "Histogram of Petal Length",
    x = "Petal Length (cm)",
    y = "Count"
  ) +
theme_minimal()
```

### 3. Box Plot - Petal.Length by Species
- Compare distributions (medians, outliers, spread).

```r
ggplot(iris, aes(x = Species, y = Petal.Length, fill = Species)) +
  geom_boxplot() +
  labs(
    title = "Petal Length by Species",
    x = "Species",
    y = "Petal Length (cm)"
  ) +
  theme_minimal()
```

### 4. Bar Chart - Count of Each Species
- Show frequency of categories.

```r
ggplot(iris, aes(x = Species, fill = Species)) +
  geom_bar() +
  labs(
    title = "Count of Each Species",
    x = "Species",
    y = "Count"
  ) +
theme_minimal()
```

#### 5. Pie Chart — Species Distribution (simulated using bar + polar)

```r
# Get counts first
species_counts <- as.data.frame(table(iris$Species))
colnames(species_counts) <- c("Species", "Count")

ggplot(species_counts, aes(x = "", y = Count, fill = Species)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  labs(
    title = "Pie Chart of Species Distribution",
    x = NULL,
    y = NULL
  ) +
  theme_void() +
  theme(legend.position = "right")
```

### 6. Line Graph - Simulated trend of Petal.Length per observation
- Line chart for trends (e.g., time series)

```r
# Simulate by arranging and adding an ID
iris_line <- iris
iris_line$ID <- 1:nrow(iris)

ggplot(iris_line, aes(x = ID, y = Petal.Length, color = Species)) +
  geom_line() +
  labs(
    title = "Simulated Line Graph of Petal Length",
    x = "Observation",
    y = "Petal Length (cm)"
  ) +
  theme_minimal()
```

### 7. Violin Plot — Petal.Width by Species
- Show full distribution shape + median

```r
ggplot(iris, aes(x = Species, y = Petal.Width, fill = Species)) +
  geom_violin(trim = FALSE) +
  labs(
    title = "Violin Plot of Petal Width by Species",
    x = "Species",
    y = "Petal Width (cm)"
  ) +
  theme_minimal()
```
